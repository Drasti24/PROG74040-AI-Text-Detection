# Train, Validation, and Test Split

This notebook divides the cleaned dataset into training, validation, and testing datasets using stratified sampling.

The same dataset partitions will be used for all machine learning models to ensure fair and consistent evaluation.

```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

## Load Clean Dataset

```python
DATA_PATH = Path("../data/processed/ai_human_text_cleaned.parquet")

df = pd.read_parquet(DATA_PATH)

df.head()
```

```python
print(df.shape)

df["generated"].value_counts(normalize=True)
```

## First split (70% train, 30% temporary)
We want 70-15-15, so first we do 70-30 and then split the 30 into 2 equal halves

```python
train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    random_state=42,
    stratify=df["generated"]
)
```

## Split the temporary 30 into validation and testing

```python
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    random_state=42,
    stratify=temp_df["generated"]
)
```

## Checking sizes of all three

```python
print("Train:", train_df.shape)
print("Validation:", val_df.shape)
print("Test:", test_df.shape)
```

## Verifying Class balance

```python
print("\nTrain")
print(train_df["generated"].value_counts(normalize=True))

print("\nValidation")
print(val_df["generated"].value_counts(normalize=True))

print("\nTest")
print(test_df["generated"].value_counts(normalize=True))
```

## Save datasets

```python
SPLITS_DIR = Path("../data/splits")
SPLITS_DIR.mkdir(parents=True, exist_ok=True)

train_df.to_parquet(SPLITS_DIR / "train.parquet", index=False)
val_df.to_parquet(SPLITS_DIR / "validation.parquet", index=False)
test_df.to_parquet(SPLITS_DIR / "test.parquet", index=False)

print("Datasets saved successfully.")
```

```python
summary = pd.DataFrame({
    "Dataset": ["Train", "Validation", "Test"],
    "Rows": [len(train_df), len(val_df), len(test_df)]
})

summary
```
